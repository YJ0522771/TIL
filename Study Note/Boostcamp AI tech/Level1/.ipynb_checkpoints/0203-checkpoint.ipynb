## 다양한 Bar Plot

```python
student = pd.read_csv('StudentsPerformance.csv')
student.sample(5) # 랜덤으로 5개, 순서대로가 아님.
```

```python
# null값 유무, 데이터의 분포, 데이터 타입 등 정보를 봄.
student.info()
```

```python
# 통계 정보
student.describe(include='all')
```

```python
# 성별에 따른 race/ethnicity 분포
group = student.groupby('gender')['race/ethnicity'].value_counts().sort_index()
display(group)
print(student['gender'].value_counts())
```

### Multiple Bar Plot

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 7), sharey=True) # sharey : 축 범위를 통일
# 축 범위를 for문으로 개별적으로 조정
# for ax in axes:
#     ax.set_ylim(0, 200) # y의 limit

axes[0].bar(group['male'].index, group['male'], color='royalblue')
axes[1].bar(group['female'].index, group['female'], color='tomato')
plt.show()
```

### Stacked Bar Plot

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 7))

group_cnt = student['race/ethnicity'].value_counts().sort_index()
axes[0].bar(group_cnt.index, group_cnt, color='darkgray') # 그룹 별 전체 수

axes[1].bar(group['male'].index, group['male'], color='royalblue')
# bottom parameter 사용.
axes[1].bar(group['female'].index, group['female'], bottom=group['male'], color='tomato') # female 데이터 밑에 male 데이터를 넣음

for ax in axes:
    ax.set_ylim(0, 350)
    
plt.show()
```

### Percentage Stacked Bar Plot 

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 7))

group = group.sort_index(ascending=False) # 역순 정렬
total=group['male']+group['female']


ax.barh(group['male'].index, group['male']/total, # 비율 계산
        color='royalblue')

ax.barh(group['female'].index, group['female']/total, 
        left=group['male']/total, # 비율 계산
        color='tomato')

ax.set_xlim(0, 1)
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False) # 각 테두리를 없앰.

plt.show()
```

### Overlapped Bar Plot

```python
group = group.sort_index() # 다시 정렬

fig, axes = plt.subplots(2, 2, figsize=(12, 12))
axes = axes.flatten()

for idx, alpha in enumerate([1, 0.7, 0.5, 0.3]): # 알파값을 변경 시키며 그려봄. (alpha = 투명도)
    axes[idx].bar(group['male'].index, group['male'], 
                  color='royalblue', 
                  alpha=alpha)
    axes[idx].bar(group['female'].index, group['female'],
                  color='tomato',
                  alpha=alpha)
    axes[idx].set_title(f'Alpha = {alpha}')
    
for ax in axes:
    ax.set_ylim(0, 200)
    
plt.show()
```

### Grouped Bar Plot

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 7))

idx = np.arange(len(group['male'].index))
width=0.35

# x축의 눈금이 좌표 0
# 왼쪽은 -좌표만큼
ax.bar(idx-width/2, group['male'], 
       color='royalblue',
       width=width, label='Male')

ax.bar(idx+width/2, group['female'], 
       color='tomato',
       width=width, label='Female')

ax.set_xticks(idx)
ax.set_xticklabels(group['male'].index)
ax.legend()    
    
plt.show()
```

```python
# 3개 이상의 그룹이 있을 때
group = student.groupby('parental level of education')['race/ethnicity'].value_counts().sort_index()
group_list = sorted(student['race/ethnicity'].unique())
edu_lv = student['parental level of education'].unique()

fig, ax = plt.subplots(1, 1, figsize=(13, 7))

x = np.arange(len(group_list))
width=0.12

for idx, g in enumerate(edu_lv):
    ax.bar(x+(-len(edu_lv)+1+2*idx)*width/2, group[g], ###### 공식
       width=width, label=g)

ax.set_xticks(x)
ax.set_xticklabels(group_list)
ax.legend()    
    
plt.show()
```

## 정확한 Bar Plot

### Principle of Proportion Ink

```python
score = student.groupby('gender').mean().T

fig, axes = plt.subplots(1, 2, figsize=(15, 7))

idx = np.arange(len(score.index))
width=0.3

for ax in axes:
    ax.bar(idx-width/2, score['male'], 
           color='royalblue',
           width=width)

    ax.bar(idx+width/2, score['female'], 
           color='tomato',
           width=width)

    ax.set_xticks(idx)
    ax.set_xticklabels(score.index)

axes[0].set_ylim(60, 75) # 0부터 시작하지 않는 축 - 직관적인 비교에 좋지 못하다. 오해를 일으킬 수 있다.
    
plt.show()
```

```python
# 비교를 위해 세로 사이즈를 키움.
fig, ax = plt.subplots(1, 1, figsize=(6, 10))

idx = np.arange(len(score.index))
width=0.3

ax.bar(idx-width/2, score['male'], 
       color='royalblue',
       width=width)

ax.bar(idx+width/2, score['female'], 
       color='tomato',
       width=width)

ax.set_xticks(idx)
ax.set_xticklabels(score.index)

plt.show()    
```

### 적절한 공간 활용

```python
group_cnt = student['race/ethnicity'].value_counts().sort_index()

fig = plt.figure(figsize=(15, 7))

ax_basic = fig.add_subplot(1, 2, 1)
ax = fig.add_subplot(1, 2, 2)

ax_basic.bar(group_cnt.index, group_cnt)
ax.bar(group_cnt.index, group_cnt,
       width=0.7, # 막대 두께
       edgecolor='black', # 테두리 색
       linewidth=2, # 테두리 선 두께
       color='royalblue'
      )

ax.margins(0.1, 0.1) # 막대 사이 공간

for s in ['top', 'right']:
    ax.spines[s].set_visible(False)

plt.show()
```

### 복잡함과 단순함

```python
group_cnt = student['race/ethnicity'].value_counts().sort_index()

fig, axes = plt.subplots(1, 2, figsize=(15, 7))

for ax in axes:
    ax.bar(group_cnt.index, group_cnt,
           width=0.7,
           edgecolor='black',
           linewidth=2,
           color='royalblue',
           zorder=10
          )

    ax.margins(0.1, 0.1)

    for s in ['top', 'right']:
        ax.spines[s].set_visible(False)

axes[1].grid(zorder=0) # grid 추가, 가독성을 고려하며 추가 여부를 결정.

for idx, value in zip(group_cnt.index, group_cnt):
    axes[1].text(idx, value+5, s=value, # 실제 값을 텍스트로 추가.
                 ha='center', 
                 fontweight='bold'
                )
        
plt.show()
```

### errorbar

```python
score_var = student.groupby('gender').std().T

fig, ax = plt.subplots(1, 1, figsize=(10, 10))

idx = np.arange(len(score.index))
width=0.3

ax.bar(idx-width/2, score['male'], 
       color='royalblue',
       width=width,
       label='Male',
       yerr=score_var['male'], # y축 오차막대
       capsize=10
      )

ax.bar(idx+width/2, score['female'], 
       color='tomato',
       width=width,
       label='Female',
       yerr=score_var['female'], # y축 오차막대
       capsize=10
      )

ax.set_xticks(idx)
ax.set_xticklabels(score.index)

ax.set_ylim(0, 100)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.legend()

ax.set_title('Gender / Score', fontsize=20)

ax.set_xlabel('Subject', fontweight='bold')
ax.set_ylabel('Score', fontweight='bold')

plt.show()
```

## 정확한 Line Plot

### 추세에 집중

```python
from matplotlib.ticker import MultipleLocator

fig = plt.figure(figsize=(12, 5))

np.random.seed(970725)

x = np.arange(20)
y = np.random.rand(20)


ax1 = fig.add_subplot(121)
ax1.plot(x, y,
         marker='o',
         linewidth=2)

# detail한 정보 표현
ax1.xaxis.set_major_locator(MultipleLocator(1)) # grid 간격 설정
ax1.yaxis.set_major_locator(MultipleLocator(0.05))    
ax1.grid(linewidth=0.3)    # grid 두께


ax2 = fig.add_subplot(122)
ax2.plot(x, y,
       linewidth=2,)

ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)

ax1.set_title(f"Line Plot (information)", loc='left', fontsize=12, va='bottom', fontweight='semibold')
ax2.set_title(f"Line Plot (clean)", loc='left', fontsize=12, va='bottom', fontweight='semibold')

plt.show()
```

### 보간

```python
from scipy.interpolate import make_interp_spline, interp1d
import matplotlib.dates as dates
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 7), dpi=300)

date_np = google.index
value_np = google['close']

date_num = dates.date2num(date_np)

# smooth
date_num_smooth = np.linspace(date_num.min(), date_num.max(), 30) # 나누는 간격
spl = make_interp_spline(date_num, value_np, k=11) # smoothing을 취하는 범위
value_np_smooth = spl(date_num_smooth)

ax[0].plot(date_np, value_np)
ax[1].plot(dates.num2date(date_num_smooth), value_np_smooth)

plt.show()
```

### 이중 축

```python
fig, ax1 = plt.subplots(figsize=(12, 7), dpi=150)

color = 'royalblue'

ax1.plot(google.index, google['close'], color=color)
ax1.set_xlabel('date')
ax1.set_ylabel('close price', color=color)  
ax1.tick_params(axis='y', labelcolor=color)

# Second Plot
# twinx 사용.
ax2 = ax1.twinx() # ax1의 x축의 이중 축

color = 'tomato'

ax2.plot(google.index, google['volume'], color=color)
ax2.set_ylabel('volume', color=color)  
ax2.tick_params(axis='y', labelcolor=color)

ax1.set_title('Google Close Price & Volume', loc='left', fontsize=15)
plt.show()
```

```python
def deg2rad(x):
    return x * np.pi / 180

def rad2deg(x):
    return x * 180 / np.pi

fig, ax = plt.subplots()

x = np.arange(0, 360)
y = np.sin(2 * x * np.pi / 180)

ax.plot(x, y)
ax.set_xlabel('angle [degrees]')
ax.set_ylabel('signal')
ax.set_title('Sine wave')

# secondary xaxis 사용
secax = ax.secondary_xaxis('top', functions=(deg2rad, rad2deg)) # 상호 변환 함수(?)
secax.set_xlabel('angle [rad]')

plt.show()
```

```python
fig = plt.figure(figsize=(12, 5))

x = np.linspace(0, 2*np.pi, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

# Ax2
ax = fig.add_subplot(111, aspect=1)
ax.plot(x, y1,
       color='#1ABDE9',
       linewidth=2,)

ax.plot(x, y2,
       color='#F36E8E',
       linewidth=2,)

ax.text(x[-1]+0.1, y1[-1], s='sin', fontweight='bold',
         va='center', ha='left', # 텍스트에서 좌표의 기준이 되는 위치
         bbox=dict(boxstyle='round,pad=0.3', fc='#1ABDE9', ec='black', alpha=0.3))

ax.text(x[-1]+0.1, y2[-1], s='cos', fontweight='bold',
         va='center', ha='left', 
         bbox=dict(boxstyle='round,pad=0.3', fc='#F36E8E', ec='black', alpha=0.3))


ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

```python
fig = plt.figure(figsize=(7, 7))

np.random.seed(97)

x = np.arange(20)
y = np.random.rand(20)

ax = fig.add_subplot(111)
ax.plot(x, y,
       color='lightgray',
       linewidth=2,)

ax.set_xlim(-1, 21)

# min과 max를 표시하는 라인을 직접 그린다.
# max
ax.plot([-1, x[np.argmax(y)]], [np.max(y)]*2,
        linestyle='--', color='tomato'
       )

ax.scatter(x[np.argmax(y)], np.max(y), 
            c='tomato',s=50, zorder=20)

# min
ax.plot([-1, x[np.argmin(y)]], [np.min(y)]*2,
        linestyle='--', color='royalblue'
       )
ax.scatter(x[np.argmin(y)], np.min(y), 
            c='royalblue',s=50, zorder=20)

plt.show()
```

# Scatter Plot 사용하기

## 기본 Scatter Plot

```python
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, aspect=1) # aspect : 가로축과 세로축의 scale. 1보다 작으면 가로축이 더 길다. 세로/가로.

np.random.seed(970725)

x = np.random.rand(20)
y = np.random.rand(20)
s = np.arange(20) * 20

ax.scatter(x, y, 
           s= s, # 크기, 데이터 수만큼 넣어주면 각 데이터에 대응하여 크기 지정. 
           c='white', # 색
           marker='o', # 마커
           linewidth=1, # 테두리 두께
           edgecolor='black') # 테두리 색

ax.set_xlim(0, 1.1) # 점이 잘리지 않도록 범위 조정
ax.set_ylim(0, 1.1)

plt.show()
```

## 정확한 Scatter Plot

```python
iris = pd.read_csv('./iris.csv')
iris.head()
```

```python
iris.describe(include='all')
```

```python
# 관계 확인을 위한 산점도
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111)

slc_mean = iris['SepalLengthCm'].mean()
swc_mean = iris['SepalWidthCm'].mean()

ax.scatter(x=iris['SepalLengthCm'], 
           y=iris['SepalWidthCm'],
           c=['royalblue' if yy <= swc_mean else 'gray' for yy in iris['SepalWidthCm']] # 조건에 따라 색을 달리함.
          )

plt.show()
```

```python
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111)

for species in iris['Species'].unique():      # label 별로 여러 번 그리기
    iris_sub = iris[iris['Species']==species]
    ax.scatter(x=iris_sub['PetalLengthCm'],   
               y=iris_sub['PetalWidthCm'],    # 지정 색이 차례로 들어감
               label=species)

ax.axvline(2.5, color='gray', linestyle=':')  # 구분 선 그리기
ax.axhline(0.8, color='gray', linestyle=':')    
ax.legend()    
plt.show()
```

```python
fig, axes = plt.subplots(4, 4, figsize=(14, 14))

# 원하는 feature들의 상관관계를 모두 그림.
feat = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

for i, f1 in enumerate(feat):
    for j, f2 in enumerate(feat):
        if i <= j : 
            axes[i][j].set_visible(False)
            continue
        for species in iris['Species'].unique():
            iris_sub = iris[iris['Species']==species]
            axes[i][j].scatter(x=iris_sub[f2], 
                               y=iris_sub[f1], 
                               label=species, 
                               alpha=0.7)   # 투명도를 줘서 밀집한 데이터도 잘 보이도록
        if i == 3: axes[i][j].set_xlabel(f2)
        if j == 0: axes[i][j].set_ylabel(f1)

plt.tight_layout()        # 여백과 관련된 서브플롯 파라미터 자동 조정.
plt.show()
```
